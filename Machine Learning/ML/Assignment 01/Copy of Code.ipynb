```python
import numpy as np
import pandas as pd
current_dir = '/content/drive/My Drive/CSC-354 Machine Learning/Assignments/01 ML'
Metadata = pd.read_csv(current_dir+'/train/Metadata.csv')
```

```python
from tensorflow import keras
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
import glob
```

```python
train_data = np.stack(
    [img_to_array(
        load_img(glob.glob(current_dir+"/train/*/"+i+"*.tiff")[0], target_size=(x, y, 3))) for i in Metadata['image'][:600]])

#Reshaping
train_data = train_data.reshape( len(train_data), x*y*3 )

#Normalizing
train_data = train_data / 255.0
```
