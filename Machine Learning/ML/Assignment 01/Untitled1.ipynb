```python
import numpy as np
import pandas as pd
Metadata = pd.read_csv('./train/Metadata.csv')
```

```python
from tensorflow import keras
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
import glob
```

```python
train_data = np.stack(
    [img_to_array(
        load_img(
            glob.glob("./train\\*\\"+i+"*.tiff")[0],
            
            target_size=(1024, 1024, 3))) for i in Metadata['image'][:2]])
```

```python
import matplotlib.pyplot as plt

plt.imshow(train_data[0], cmap=plt.cm.binary)
plt.show()
```
